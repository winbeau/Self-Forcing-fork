<cell_type>markdown</cell_type># Figure 4 Reproduction: Temporal Attention Visualization

This notebook reproduces Figure 4 from the paper, showing how attention weights
are distributed across frames during autoregressive video generation.

**Paper's Key Finding:**
> "Contrary to the conventional understanding that only a small set of initial KV tokens 
> (latent frames) needs to be retained, our analysis reveals: most attention heads allocate 
> substantial weight to not only the earliest tokens, but also assign comparable attention 
> to the middle of the sequence."

**Figure 4 Description:**
> "Query-averaged attention showing how the last chunk (frames 19-21) attends to earlier 
> KV cache entries (frames 0-18). We visualize two representative attention heads from 
> different layers—L1H1 (layer 1, head 1) and L5H10 (layer 5, head 10)."

````python
# Inspect the captured attention weights
print("\nCaptured attention weight tensors:")
print("=" * 80)
for i, w in enumerate(attention_weights):
    lq, lk = w['attn_weights'].shape[2], w['attn_weights'].shape[3]
    num_heads = w['attn_weights'].shape[1]
    q_frames = lq // frame_seq_length
    k_frames = lk // frame_seq_length
    
    # Determine attention type
    if lk == lq:
        attn_type = "Full self-attention (current block)"
    elif lk < frame_seq_length:
        attn_type = f"Sliding window KV cache ({lk} tokens)"
    else:
        attn_type = f"KV cache attention ({k_frames} frames)"
    
    print(f"  [{i}] Layer {w['layer_idx']:2d}: "
          f"shape={w['attn_weights'].shape}, "
          f"Q={q_frames} frames, K={k_frames if k_frames > 0 else f'{lk} tokens'}")
    print(f"       -> {attn_type}")
print("=" * 80)

# Find layers with full KV cache for Figure 4 analysis
full_kv_layers = [i for i, w in enumerate(attention_weights) 
                  if w['attn_weights'].shape[3] == w['attn_weights'].shape[2]]
print(f"\nLayers with full attention matrix (for Figure 4): {[attention_weights[i]['layer_idx'] for i in full_kv_layers]}")
````

## 2. Process Attention Weights

The attention tensor shape is typically `(Batch, Heads, Query_Len, Key_Len)`.

For Figure 4, we need to:
1. Select a specific layer and head
2. Average across the Query dimension to get the "average attention per key position"
3. Map token positions to frame indices

````python
def compute_frame_attention_distribution(
    attn_weights: torch.Tensor,
    frame_seq_length: int = 1560,
    query_frame_range: tuple = None,
    key_frame_range: tuple = None,
):
    """
    Compute attention distribution across key frames for Figure 4.
    
    For Figure 4, we want to see how the last chunk (query frames 19-21) 
    attends to earlier KV cache entries (key frames 0-18).
    
    Args:
        attn_weights: Attention tensor of shape [B, num_heads, Lq, Lk]
        frame_seq_length: Number of tokens per frame
        query_frame_range: (start, end) frame indices for query (default: last 3 frames)
        key_frame_range: (start, end) frame indices for key (default: all earlier frames)
    
    Returns:
        frame_attention: Array of shape [num_heads, num_key_frames] 
                        showing average attention per key frame for each head
        key_frame_indices: Array of key frame indices
    """
    B, num_heads, lq, lk = attn_weights.shape
    
    # Calculate frame counts
    num_query_frames = lq // frame_seq_length
    num_key_frames = lk // frame_seq_length
    
    # Handle case where key length < frame_seq_length (sliding window)
    if num_key_frames == 0:
        # For sliding window, group tokens into pseudo-frames
        tokens_per_group = frame_seq_length // 10  # ~156 tokens per group
        num_key_frames = max(1, lk // tokens_per_group)
        frame_seq_length = tokens_per_group
    
    # Default: query is all frames, key is all frames
    if query_frame_range is None:
        query_frame_range = (0, num_query_frames)
    if key_frame_range is None:
        key_frame_range = (0, num_key_frames)
    
    # Extract attention matrix [B, num_heads, Lq, Lk] -> [num_heads, Lq, Lk]
    attn = attn_weights[0].float().numpy()  # Take batch 0
    
    # Compute average attention per key frame for each head
    # Average over query positions to get "how much attention each key position receives"
    frame_attention = np.zeros((num_heads, num_key_frames))
    
    for head_idx in range(num_heads):
        head_attn = attn[head_idx]  # [Lq, Lk]
        
        # Average over all query positions
        avg_attn_per_key_token = head_attn.mean(axis=0)  # [Lk]
        
        # Group by key frame
        for kf in range(num_key_frames):
            k_start = kf * frame_seq_length
            k_end = min((kf + 1) * frame_seq_length, lk)
            if k_end > k_start:
                frame_attention[head_idx, kf] = avg_attn_per_key_token[k_start:k_end].mean()
    
    key_frame_indices = np.arange(num_key_frames)
    
    return frame_attention, key_frame_indices
````

````python
# Compute attention distributions for all captured layers
layer_attention_data = {}

for i, w in enumerate(attention_weights):
    layer_idx = w['layer_idx']
    attn = w['attn_weights']
    
    frame_attention, key_frame_indices = compute_frame_attention_distribution(
        attn, frame_seq_length=frame_seq_length
    )
    
    layer_attention_data[layer_idx] = {
        'frame_attention': frame_attention,  # [num_heads, num_frames]
        'key_frame_indices': key_frame_indices,
        'attn_shape': attn.shape,
        'num_heads': attn.shape[1],
    }
    
    print(f"Layer {layer_idx}: {frame_attention.shape[1]} key frames, {frame_attention.shape[0]} heads")
    print(f"  Attention range: [{frame_attention.min():.6f}, {frame_attention.max():.6f}]")
````

## 3. Plot Figure 4: Attention Distribution Across Frames

````python
def plot_figure4_paper_style(
    layer_attention_data: dict,
    heads_to_show: list = None,
    save_path: str = None,
    figsize: tuple = (14, 5),
):
    """
    Reproduce Figure 4 from the paper showing attention distribution across frames.
    
    Paper description:
    "Query-averaged attention showing how the last chunk (frames 19-21) attends to 
    earlier KV cache entries (frames 0-18). We visualize two representative attention 
    heads from different layers—L1H1 (layer 1, head 1) and L5H10 (layer 5, head 10)—
    demonstrating that substantial attention is maintained across the entire context 
    window, not just initial frames."
    
    Args:
        layer_attention_data: Dict with layer_idx -> attention data
        heads_to_show: List of (layer_idx, head_idx, label) tuples
        save_path: Path to save SVG
        figsize: Figure size
    """
    # Default: show representative heads from different layers
    if heads_to_show is None:
        # Try to match paper's L1H1 and L5H10
        heads_to_show = []
        if 0 in layer_attention_data:
            heads_to_show.append((0, 0, "L1H1"))  # Layer 1 (0-indexed=0), Head 1 (0-indexed=0)
        if 1 in layer_attention_data:
            heads_to_show.append((1, 0, "L2H1"))
        if 4 in layer_attention_data:
            heads_to_show.append((4, 9, "L5H10"))  # Layer 5 (0-indexed=4), Head 10 (0-indexed=9)
        if 5 in layer_attention_data:
            heads_to_show.append((5, 9, "L6H10"))
        if 14 in layer_attention_data:
            heads_to_show.append((14, 0, "L15H1"))
    
    num_plots = len(heads_to_show)
    if num_plots == 0:
        print("No valid heads to show!")
        return None
    
    # Use 2 columns for paper-style layout
    ncols = min(2, num_plots)
    nrows = (num_plots + ncols - 1) // ncols
    
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize, squeeze=False)
    axes = axes.flatten()
    
    # Color palette
    colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#3B1F2B']
    
    for idx, (layer_idx, head_idx, label) in enumerate(heads_to_show):
        ax = axes[idx]
        
        if layer_idx not in layer_attention_data:
            ax.set_visible(False)
            continue
        
        data = layer_attention_data[layer_idx]
        frame_attention = data['frame_attention']
        key_frame_indices = data['key_frame_indices']
        
        # Get attention for this specific head
        if head_idx >= frame_attention.shape[0]:
            head_idx = 0  # Fallback to first head
        
        head_attn = frame_attention[head_idx]
        
        # Normalize for better visualization (optional)
        # head_attn = head_attn / head_attn.sum() if head_attn.sum() > 0 else head_attn
        
        color = colors[idx % len(colors)]
        
        # Plot bar chart style (more similar to paper)
        ax.bar(key_frame_indices, head_attn, color=color, alpha=0.7, edgecolor=color, linewidth=0.5)
        
        # Also add line plot for trend
        ax.plot(key_frame_indices, head_attn, 'o-', color=color, linewidth=1.5, markersize=4, alpha=0.9)
        
        # Labels
        ax.set_xlabel('Key Frame Index', fontsize=11)
        ax.set_ylabel('Attention Weight', fontsize=11)
        ax.set_title(f'{label} (Layer {layer_idx+1}, Head {head_idx+1})', fontsize=12, fontweight='bold')
        
        # Grid
        ax.grid(True, alpha=0.3, linestyle='--')
        ax.set_axisbelow(True)
        
        # X-axis ticks
        if len(key_frame_indices) <= 10:
            ax.set_xticks(key_frame_indices)
        else:
            # Show every few frames
            step = max(1, len(key_frame_indices) // 10)
            ax.set_xticks(key_frame_indices[::step])
        
        # Add statistics annotation
        max_idx = np.argmax(head_attn)
        ax.annotate(f'max@{key_frame_indices[max_idx]}', 
                   xy=(key_frame_indices[max_idx], head_attn[max_idx]),
                   xytext=(5, 5), textcoords='offset points', fontsize=8, color='darkgreen')
    
    # Hide unused subplots
    for idx in range(num_plots, len(axes)):
        axes[idx].set_visible(False)
    
    plt.suptitle(
        'Figure 4: Attention Weight Distribution Across Earlier Frames\n'
        '(Query-averaged attention from last chunk to KV cache)',
        fontsize=13, fontweight='bold', y=1.02
    )
    plt.tight_layout()
    
    if save_path:
        svg_path = save_path if save_path.endswith('.svg') else save_path.replace('.png', '.svg')
        plt.savefig(svg_path, format='svg', bbox_inches='tight', dpi=150)
        print(f"Saved to {svg_path}")
    
    plt.show()
    return fig
````

````python
# Plot Figure 4 - Paper style with two representative heads
# According to the paper: L1H1 (layer 1, head 1) and L5H10 (layer 5, head 10)

# Only use layers with full attention matrix (not sliding window truncated)
fig = plot_figure4_paper_style(
    layer_attention_data,
    heads_to_show=[
        (0, 0, "L1H1"),   # Layer 1, Head 1
        (4, 9, "L5H10"),  # Layer 5, Head 10
    ],
    save_path="figure4_reproduction.svg",
    figsize=(12, 4)
)
````

## 4. Multi-Head Analysis

Let's visualize attention patterns across different heads to see if they show different behaviors.

````python
def plot_all_heads_heatmap(
    layer_attention_data: dict,
    layer_idx: int,
    save_path: str = None,
    figsize: tuple = (14, 6),
):
    """
    Plot attention distribution for all heads in a layer as a heatmap.
    """
    if layer_idx not in layer_attention_data:
        print(f"Layer {layer_idx} not found in data!")
        return None
    
    data = layer_attention_data[layer_idx]
    frame_attention = data['frame_attention']  # [num_heads, num_frames]
    key_frame_indices = data['key_frame_indices']
    num_heads = frame_attention.shape[0]
    
    fig, ax = plt.subplots(figsize=figsize)
    
    # Create heatmap
    im = ax.imshow(frame_attention, cmap='viridis', aspect='auto', interpolation='nearest')
    plt.colorbar(im, ax=ax, label='Attention Weight')
    
    # Labels
    ax.set_xlabel('Key Frame Index', fontsize=12)
    ax.set_ylabel('Attention Head', fontsize=12)
    ax.set_title(f'Layer {layer_idx + 1}: Attention Distribution Across All Heads', 
                fontsize=14, fontweight='bold')
    
    # Y-axis: head indices
    ax.set_yticks(range(num_heads))
    ax.set_yticklabels([f'H{i+1}' for i in range(num_heads)])
    
    # X-axis: frame indices
    if len(key_frame_indices) <= 10:
        ax.set_xticks(range(len(key_frame_indices)))
        ax.set_xticklabels(key_frame_indices)
    else:
        step = max(1, len(key_frame_indices) // 10)
        ax.set_xticks(range(0, len(key_frame_indices), step))
        ax.set_xticklabels(key_frame_indices[::step])
    
    plt.tight_layout()
    
    if save_path:
        svg_path = save_path if save_path.endswith('.svg') else save_path.replace('.png', '.svg')
        plt.savefig(svg_path, format='svg', bbox_inches='tight')
        print(f"Saved to {svg_path}")
    
    plt.show()
    return fig
````

````python
# Plot all heads heatmap for layers with full attention
# Layer 0 (L1) - full attention
if 0 in layer_attention_data:
    fig = plot_all_heads_heatmap(
        layer_attention_data,
        layer_idx=0,
        save_path="figure4_layer1_all_heads.svg"
    )

# Layer 4 (L5) - full attention  
if 4 in layer_attention_data:
    fig = plot_all_heads_heatmap(
        layer_attention_data,
        layer_idx=4,
        save_path="figure4_layer5_all_heads.svg"
    )
````

## 5. Full Attention Matrix Heatmap

Visualize the complete attention matrix to understand Query-Key relationships.

````python
def plot_attention_comparison(
    layer_attention_data: dict,
    layers_to_compare: list = None,
    save_path: str = None,
    figsize: tuple = (14, 5),
):
    """
    Compare head-averaged attention distributions across different layers.
    This helps verify the paper's claim that attention is distributed across
    the entire context window, not just initial frames.
    """
    if layers_to_compare is None:
        layers_to_compare = sorted(layer_attention_data.keys())
    
    fig, ax = plt.subplots(figsize=figsize)
    
    colors = plt.cm.viridis(np.linspace(0, 1, len(layers_to_compare)))
    
    for idx, layer_idx in enumerate(layers_to_compare):
        if layer_idx not in layer_attention_data:
            continue
        
        data = layer_attention_data[layer_idx]
        frame_attention = data['frame_attention']  # [num_heads, num_frames]
        key_frame_indices = data['key_frame_indices']
        
        # Average across all heads
        mean_attention = frame_attention.mean(axis=0)
        std_attention = frame_attention.std(axis=0)
        
        color = colors[idx]
        ax.plot(key_frame_indices, mean_attention, 'o-', 
               color=color, linewidth=2, markersize=5, 
               label=f'Layer {layer_idx + 1}', alpha=0.8)
        ax.fill_between(key_frame_indices, 
                       mean_attention - std_attention,
                       mean_attention + std_attention,
                       color=color, alpha=0.2)
    
    ax.set_xlabel('Key Frame Index', fontsize=12)
    ax.set_ylabel('Average Attention Weight', fontsize=12)
    ax.set_title('Head-Averaged Attention Distribution Across Layers', 
                fontsize=14, fontweight='bold')
    ax.legend(loc='upper right', fontsize=10)
    ax.grid(True, alpha=0.3, linestyle='--')
    
    plt.tight_layout()
    
    if save_path:
        svg_path = save_path if save_path.endswith('.svg') else save_path.replace('.png', '.svg')
        plt.savefig(svg_path, format='svg', bbox_inches='tight')
        print(f"Saved to {svg_path}")
    
    plt.show()
    return fig
````

````python
# Compare attention distributions across layers (only full attention layers)
full_attn_layers = [l for l in layer_attention_data.keys() 
                    if layer_attention_data[l]['frame_attention'].shape[1] == 3]  # 3 frames = full block

fig = plot_attention_comparison(
    layer_attention_data,
    layers_to_compare=full_attn_layers,
    save_path="figure4_layer_comparison.svg"
)
````

## 6. Head-Averaged Attention

Average across all heads for a more robust visualization.

````python
def plot_attention_statistics(layer_attention_data: dict, save_path: str = None):
    """
    Print and visualize statistics about the attention distribution.
    Verify the paper's claim about attention being distributed across entire context.
    """
    print("=" * 70)
    print("ATTENTION ANALYSIS SUMMARY (Figure 4 Reproduction)")
    print("=" * 70)
    
    # Collect statistics for layers with full attention
    stats_data = []
    
    for layer_idx in sorted(layer_attention_data.keys()):
        data = layer_attention_data[layer_idx]
        frame_attention = data['frame_attention']
        num_frames = frame_attention.shape[1]
        
        # Skip sliding window layers for main analysis
        if num_frames < 3:
            continue
        
        # Head-averaged attention
        mean_attn = frame_attention.mean(axis=0)
        
        # Statistics
        first_frame_attn = mean_attn[0]
        last_frame_attn = mean_attn[-1]
        middle_attn = mean_attn[1:-1].mean() if len(mean_attn) > 2 else mean_attn.mean()
        max_frame = np.argmax(mean_attn)
        min_frame = np.argmin(mean_attn)
        
        stats_data.append({
            'layer': layer_idx + 1,
            'first': first_frame_attn,
            'middle': middle_attn,
            'last': last_frame_attn,
            'max_frame': max_frame,
            'min_frame': min_frame,
            'ratio_first_middle': first_frame_attn / middle_attn if middle_attn > 0 else 0,
        })
        
        print(f"\nLayer {layer_idx + 1}:")
        print(f"  First frame attention:  {first_frame_attn:.6f}")
        print(f"  Middle frames attention: {middle_attn:.6f}")
        print(f"  Last frame attention:   {last_frame_attn:.6f}")
        print(f"  Max attention at frame: {max_frame}")
        print(f"  First/Middle ratio:     {first_frame_attn/middle_attn:.2f}x" if middle_attn > 0 else "N/A")
    
    print("\n" + "=" * 70)
    print("KEY FINDING (from paper):")
    print("  'Most attention heads allocate substantial weight to not only")
    print("   the earliest tokens, but also assign comparable attention to")
    print("   the middle of the sequence.'")
    print("=" * 70)
    
    # Create summary visualization
    if len(stats_data) > 0:
        fig, ax = plt.subplots(figsize=(10, 5))
        
        layers = [s['layer'] for s in stats_data]
        first_attns = [s['first'] for s in stats_data]
        middle_attns = [s['middle'] for s in stats_data]
        last_attns = [s['last'] for s in stats_data]
        
        x = np.arange(len(layers))
        width = 0.25
        
        ax.bar(x - width, first_attns, width, label='First Frame', color='#2E86AB')
        ax.bar(x, middle_attns, width, label='Middle Frames', color='#F18F01')
        ax.bar(x + width, last_attns, width, label='Last Frame', color='#A23B72')
        
        ax.set_xlabel('Layer', fontsize=12)
        ax.set_ylabel('Average Attention Weight', fontsize=12)
        ax.set_title('Attention Distribution: First vs Middle vs Last Frames', 
                    fontsize=14, fontweight='bold')
        ax.set_xticks(x)
        ax.set_xticklabels([f'L{l}' for l in layers])
        ax.legend()
        ax.grid(True, alpha=0.3, axis='y')
        
        plt.tight_layout()
        
        if save_path:
            svg_path = save_path if save_path.endswith('.svg') else save_path.replace('.png', '.svg')
            plt.savefig(svg_path, format='svg', bbox_inches='tight')
            print(f"\nSaved to {svg_path}")
        
        plt.show()
        return fig
    
    return None
````

````python
# Generate statistics and summary visualization
fig = plot_attention_statistics(layer_attention_data, save_path="figure4_statistics.svg")
````

<cell_type>markdown</cell_type>## 8. Notes

### Interpreting Figure 4 Results

**Paper's Key Finding:**
> "Contrary to the conventional understanding that only a small set of initial KV tokens (latent frames) needs to be retained, our analysis reveals: most attention heads allocate substantial weight to not only the earliest tokens, but also assign comparable attention to the middle of the sequence."

### What This Means:

1. **Not Just Attention Sinks**: Unlike LLMs where attention often concentrates on initial tokens (attention sink pattern), video diffusion models distribute attention more evenly across all frames.

2. **Full Context Matters**: For autoregressive video generation, retaining the full KV cache is important - not just the first few frames and recent frames.

3. **Layer Differences**: Different layers may show different attention patterns:
   - Some layers show full self-attention within the current block
   - Some layers show sliding window attention to the KV cache

### Technical Notes:

- **Query**: Last chunk (frames 19-21 in a 21-frame video)
- **Key**: Earlier KV cache entries (frames 0-18)
- **Attention shape**: `[B, num_heads, Lq, Lk]`
- **Frame seq length**: 1560 tokens per frame (60×104 / 4 patches)

### Differences from Original Figure 4:

- The exact attention pattern depends on:
  - The specific checkpoint used
  - The prompt content
  - The random seed
- Different layers have different attention window sizes
- The paper shows specific representative heads (L1H1, L5H10)
